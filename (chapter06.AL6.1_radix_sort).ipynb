2) 알고리즘 설명 (MD)

기수 정렬(Radix Sort) 알고리즘은 비교 정렬이 아닌 정렬 알고리즘으로, 정수들을 자릿수 별로 비교하여 정렬합니다. 기수 정렬은 각 자리의 숫자를 기준으로 정렬하는데, 이는 각 숫자의 범위가 작은 경우에 매우 효율적입니다.

알고리즘은 다음과 같은 과정을 거칩니다:

가장 낮은 자릿수부터 시작하여 가장 높은 자릿수까지 반복합니다.
각 자릿수에 대해 정렬을 수행합니다.
각 자릿수의 값에 따라 숫자들을 버킷에 나누어 담습니다.
버킷에 담긴 숫자들을 순서대로 다시 합칩니다.

## 복잡도 분석

### 기수 정렬 알고리즘의 복잡도

기수 정렬 알고리즘의 시간 복잡도는 입력 크기에 따라 다릅니다. 여기서 입력 크기는 정렬할 정수의 개수를 나타냅니다.

- **최선, 평균, 최악의 경우 모두 O(k * n)**: 여기서 k는 최대 자릿수를 나타냅니다.
- **공간 복잡도는 O(n + k)**: 여기서 n은 입력 배열의 크기를 나타내며, k는 버킷의 수를 나타냅니다.

### 복잡도 분석

- 입력으로 받은 리스트의 최대 자릿수를 구하는 과정에 O(n)의 시간이 소요됩니다.
- 최대 자릿수를 구한 후, 각 자릿수별로 기수 정렬을 수행하면서 전체적인 정렬을 진행합니다. 이 과정에서 각 자릿수별로 정렬하는 데 O(n)의 시간이 필요합니다. 자릿수는 보통 상수개이므로 이 과정은 O(k * n)의 시간이 소요됩니다.

따라서 기수 정렬 알고리즘은 입력 크기와 무관하게 상수 시간 내에 실행되는 효율적인 정렬 알고리즘 중 하나입니다.
