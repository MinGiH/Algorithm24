1) 문제 정의 (MD)

1- 기수 정렬
기수 정렬은 정렬할 항목들의 자릿수를 기준으로 정렬을 수행하는 알고리즘이다. 이 알고리즘은 주로 정수나 고정 길이의 문자열 같은 경우에 사용되며, 비교 기반 정렬 알고리즘과는 다르게 자릿수별로 처리를 함으로써 효율적인 정렬을 가능하게 한다.

2- 카운팅 정렬
카운팅 정렬은 비교 기반 정렬이 아닌 정렬 알고리즘으로, 주어진 배열의 각 요소가 정수이고, 그 범위가 제한된 경우에 사용된다. 이 알고리즘은 정수의 빈도를 세고, 그 빈도에 따라 정렬을 수행하여 효율적으로 데이터를 정렬한다.

3- 문자열매칭(호어 알고리즘) 
호어 알고리즘은 문자열 매칭 문제에서 패턴을 본문에서 찾기 위해 사용되는 효율적인 알고리즘이다. 이 알고리즘은 일반적으로 빠른 검색을 위해 개발된 알고리즘 중 하나로, 특히 부분 문자열을 찾는 데 유용하다.

4- 해시
해시는 데이터를 고정된 크기의 해시 값으로 변환하는 함수로, 주로 데이터 검색과 저장에서 효율성을 극대화하기 위해 사용된다. 해시는 다양한 분야에서 사용되며, 여기서는 주어진 데이터를 해시 테이블에 저장하고 검색하는 문제를 정의한다.

 2) 알고리즘 설명  (MD)

 1- 기수 정렬 
 1. 가장 낮은 자릿수부터 시작해 각 자릿수를 기준으로 배열을 정렬한다.
 2. 정렬은 안정 정렬(예: 계수 정렬)을 사용한다.
 3. 모든 자릿수에 대해 정렬이 완료되면 배열이 정렬된 상태가 된다.


2- 카운팅 정렬 
1. 배열의 각 값의 빈도를 세어 count 배열에 저장한다.
2. count 배열을 사용해 누적 합을 계산한다.
3. 누적 합을 사용해 각 값을 올바른 위치에 배치해 정렬된 배열을 생성한다.

3- 문자열 매칭 (호어 알고리즘)
1.패턴의 각 문자에 대한 오른쪽으로부터의 위치를 기록하는 테이블을 만든다.
2.본문 문자열을 패턴 문자열과 비교하며 일치 여부를 검사한다.
3.불일치가 발생하면 테이블을 참조해 패턴을 오른쪽으로 이동시킨다.

4- 해시
1.해시 함수를 사용해 키를 해시 값으로 변환하여 인덱스를 생성한다.
2.충돌이 발생하면 개방 주소법(빈 슬롯을 찾음) 또는 체이닝(연결 리스트 사용)을 통해 해결한다.
3.검색 시 해시 값을 계산해 인덱스를 찾고, 해당 위치의 값을 반환한다.

4) 코드 개요(입력 변수, 출력, 함수 설명) (MD)

1- 기수 정렬
입력 변수: arr: 정렬할 정수 배열.
출력: 정렬된 배열 (정렬된 후 원래 배열을 수정한다).
함수 설명: radix_sort(arr): 주어진 배열을 기수 정렬하는 함수

2- 카운팅 정렬
입력 변수: arr: 정렬할 정수 배열.
출력: 정렬된 배열 (정렬된 후 원래 배열을 수정한다).
함수 설명: counting_sort(arr): 주어진 배열을 카운팅 정렬하는 함수.

3- 문자열 매칭 (호어 알고리즘)
입력 변수: T: 본문 문자열.
P: 패턴 문자열.
출력: 패턴이 나타나는 시작 위치의 리스트.
함수 설명: hoare_preprocess(P): 패턴의 각 문자에 대한 오른쪽 위치 테이블을 생성하는 함수.
hoare_search(T, P): 본문에서 패턴을 검색하는 함수.

4- 해시
입력 변수: key: 키 값.
value: 저장할 값.
출력: 검색된 값 또는 "Not Found".
함수 설명: hash_function(key): 키를 해시 값으로 변환하는 함수.
insert(key, value): 키-값 쌍을 해시 테이블에 삽입하는 함수.
search(key): 해시 테이블에서 키를 검색하는 함수.

```python
def hoare_preprocess(P):
    m = len(P)
    shift = [m] * 256
    for i in range(m - 1):
        shift[ord(P[i])] = m - 1 - i
    return shift

def hoare_search(T, P):
    n, m = len(T), len(P)
    if m > n:
        return []

    shift = hoare_preprocess(P)
    result = []
    s = 0

    while s <= n - m:
        j = m - 1
        while j >= 0 and P[j] == T[s + j]:
            j -= 1
        if j < 0:
            result.append(s)
            s += shift[ord(T[s + m])] if s + m < n else 1
        else:
            s += max(1, shift[ord(T[s + j])])
    
    return result

```

```python
class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    def hash_function(self, key):
        return hash(key) % self.size

    def insert(self, key, value):
        index = self.hash_function(key)
        original_index = index
        while self.table[index] is not None:
            if self.table[index][0] == key:
                self.table[index] = (key, value)
                return
            index = (index + 1) % self.size
            if index == original_index:
                raise Exception("Hash table is full")
        self.table[index] = (key, value)

    def search(self, key):
        index = self.hash_function(key)
        original_index = index
        while self.table[index] is not None:
            if self.table[index][0] == key:
                return self.table[index][1]
            index = (index + 1) % self.size
            if index == original_index:
                break
        return "Not Found"

```

```python
from queue import Queue

BUCKETS = 10
DIGITS = 4

def radix_sort(A):
    queues = []
    for _ in range(BUCKETS):
        queues.append(Queue())

    n = len(A)
    factor = 1
    for d in range(DIGITS):
        for i in range(n):
            queues[(A[i] // factor) % BUCKETS].put(A[i])
        j = 0
        for b in range(BUCKETS):
            while not queues[b].empty():
                A[j] = queues[b].get()
                j += 1
        factor *= BUCKETS
        print("step", d+1, A)

# Test Cases
test_cases = [
    [170, 45, 75, 90, 802, 24, 2, 66],
    [3, 6, 9, 1, 2, 0, 8, 7, 5, 4],
    [5, 4, 3, 2, 1],
    [12, 34, 54, 2, 3],
    [123, 456, 789, 101, 102]
]

for i, arr in enumerate(test_cases):
    print("Original:", arr)
    radix_sort(arr)
    print("Sorted:", arr)
    print()

```

```python
def counting_sort(A):
    MAX_VAL = max(A) + 1
    output = [0] * len(A)
    count = [0] * MAX_VAL

    for i in A:
        count[i] += 1

    for i in range(1, MAX_VAL):
        count[i] += count[i - 1]

    for i in range(len(A) - 1, -1, -1):
        output[count[A[i]] - 1] = A[i]
        count[A[i]] -= 1

    for i in range(len(A)):
        A[i] = output[i]

# Test Cases
test_cases = [
    [4, 2, 2, 8, 3, 3, 1],
    [10, 7, 8, 9, 1, 5],
    [1, 4, 1, 2, 7, 5, 2],
    [5, 3, 8, 6, 2, 7, 4, 1],
    [3, 6, 9, 2, 7, 1, 8]
]

for i, arr in enumerate(test_cases):
    print("Original:", arr)
    counting_sort(arr)
    print("Sorted:", arr)
    print()
```

1. 기수 정렬 (Radix Sort):
시간 복잡도: Best Case: O(n * k)
Average Case: O(n * k)
Worst Case: O(n * k)
여기서 n은 배열의 요소 수이고, k는 배열 요소의 최대 자릿수이다.
공간 복잡도: O(n + k)

2. 카운팅 정렬 (Counting Sort):
시간 복잡도:
Best Case: O(n + k)
Average Case: O(n + k)
Worst Case: O(n + k)
여기서 n은 배열의 요소 수이고, k는 배열 요소의 최대 값이다.
공간 복잡도:
O(n + k)

3. 문자열 매칭 (호어 알고리즘):
시간 복잡도:
Best Case: O(n + m)
Average Case: O(n + m)
Worst Case: O(n * m)
여기서 n은 텍스트의 길이이고, m은 패턴의 길이이다.
공간 복잡도:
O(m)

4. 해시 (Hash):
시간 복잡도:
삽입: O(1) (평균적으로)
탐색: O(1) (평균적으로)
공간 복잡도:
O(n) (n은 해시 테이블의 크기)
